```python
from sklearn import linear_model # preparing a linear model

clf = linear_model.LogisticRegression() # building a classifier
```

```python
clf
```

```python
# Importing functions to split a dataset
from sklearn.model_selection import train_test_split

# Spliting into train & test set, here the train set is 70% and test set is 30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
import warnings # hide 'lbfgs' warning
warnings.filterwarnings('ignore')
```

```python
clf.fit(X_train, y_train); # Training the model by classifier. Use the function imported above and apply fit() on it.
```

```python
clf.score(X_train, y_train) # accuracy of predictions for training data
```

```python
clf.score(X_test, y_test) # accuracy of predictions for test data
```

```python
clf.predict(X_test) # Prediction outputs for test data. Using the predict() on the model to predict the output.
```

```python
# Predicted outputs of test data and the original test data correct or incorrect.
a = clf.predict(X_test)
b = y_test

# Collating data from predicted outputs of test data and the original test data.
pd.set_option('display.max_rows', 200)
pd.DataFrame(a == b)
```

```python
df = pd.DataFrame({'A': a == b})
print(df['A'].value_counts())
```
